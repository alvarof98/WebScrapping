```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
import numpy as np
from time import sleep
import random
```

```python
navegador = webdriver.Chrome('chromedriver.exe')

navegador.get('https://www.carrefour.es/supermercado/productos-frescos/cheeses/cat20020/c')
```

```python
lista_total = []
while True:
    try:
        sleep(5)
        navegador.find_element_by_class_name('pagination__main').click()
        sleep(5)
        soup = BeautifulSoup(navegador.page_source)
        lista_productos = soup.find('div',{'class':'product-card-list'}).find_all('li')
        print(len(lista_productos))
        for i in lista_productos:
            try:
                diccionario = {}
                diccionario['producto'] = i.find(attrs={'class':'link product-card__title-link track-click'}).text.strip('\n          ').strip('.\n        ')
                diccionario['precio'] = i.find(attrs = {'class': 'product-card__parent'})['app_price']
                diccionario['precio por kilo'] = i.find(attrs = {'class': 'product-card__parent'})['app_price_per_unit']
                lista_total.append(diccionario)
                print('Hola: ',len(lista_total))
            except:
                continue
        if soup.find('span',{'aria-label':'Página siguiente'}) is not None:
            sleep(5)
            navegador.find_element_by_class_name('pagination__next').click()
        else:
            break 
    except:
        try:
            navegador.find_element_by_id('onetrust-accept-btn-handler').click()
        except:
            break
```

```python
import json
with open('quesos.json', 'w') as outfile:
    json.dump(lista_total, outfile)
```
