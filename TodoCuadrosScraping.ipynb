# DESTRIPANDO CUADRO A CUADRO

```python
def info_cuadro(link):
    reponse = requests.get(link)
    soup = BeautifulSoup(reponse.text, 'html.parser')
    sub_soup = soup.find(attrs = {'class':'relacionados'}).find_all('div')
    lista = ['Autor','Estilo','Tema','Título original','Título (inglés)','Tipo','Técnica','Soporte','Año','Se encuentra en','Género','Está en']
    diccionario = {}
    for i in lista:
        for j in sub_soup:
            if i in j.text:
                diccionario[i] = j.text.split(': ')[-1]
                break
            else:
                diccionario[i] = np.nan
    return diccionario
```
